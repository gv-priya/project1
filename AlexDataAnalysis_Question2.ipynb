```python
##Alex, Question 2 
##Question###
#Question 2: Within the calendar year what is the best month/season to set up the retail business?

#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
##############Data Source 2########################

#Read data source, 
#drop additional header rows. 
#dipsplay head ru
data_by_state = pd.read_excel('Data/Table 3. Retail turnover, by state 850103.xlsx',sheet_name="Data1")
data_by_state = data_by_state.drop(range(0,9))
data_by_state = data_by_state.rename(columns = {'Unnamed: 0':'Month'})
data_by_state.head()
```

```python
#create a list of states to group by. 
state_list = ['New South Wales', 'Victoria', 'Queensland', 'South Australia', 'Western Australia', 'Tasmania', 'Northern Territory', 'Australian Capital Territory']
```

```python

#create data frame and lock the columnms for raw data totals
question_2_df = data_by_state.loc[:,["Month",
                            "Turnover ;  New South Wales ;  Total (Industry) ;",
                            "Turnover ;  Victoria ;  Total (Industry) ;",
                            "Turnover ;  Queensland ;  Total (Industry) ;",
                            "Turnover ;  South Australia ;  Total (Industry) ;",
                            "Turnover ;  Western Australia ;  Total (Industry) ;",
                            "Turnover ;  Tasmania ;  Total (Industry) ;",
                            "Turnover ;  Northern Territory ;  Total (Industry) ;",
                            "Turnover ;  Australian Capital Territory ;  Total (Industry) ;"                  
                                    ]]

chosen_state_df = question_2_df.filter(regex= 'Month|')
```

```python
#set a loop for 
for state in state_list:

    #set the state to the current item in the loop
    chosen_state = state
    
    
    #filter data to state m
    chosen_state_df = question_2_df.filter(regex= f'Month|{chosen_state}')
    chosen_state_df = chosen_state_df.set_index('Month')


    #filter to the relevnat 
    chosen_state_last_decade_df = chosen_state_df.loc['2013-01-01':'2022-12-31']


    #specify line plot for graph
    chosen_state_last_decade_df.plot(kind='line')

    
    plt.title(f"Industry Turnover by Month {chosen_state} ")
    plt.xlabel('Month')
    plt.ylabel('Turnover ($ Millions)')
    plt.legend()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07),
          fancybox=True, shadow=True, ncol=5)

plt.show()
```

```python

#change dataframe to specify seasonally adjusted
question_2_df = data_by_state.loc[:,["Month",
                            "Turnover ;  New South Wales ;  Total (Industry) ;.1",
                            "Turnover ;  Victoria ;  Total (Industry) ;.1",
                            "Turnover ;  Queensland ;  Total (Industry) ;.1",
                            "Turnover ;  South Australia ;  Total (Industry) ;.1",
                            "Turnover ;  Western Australia ;  Total (Industry) ;.1",
                            "Turnover ;  Tasmania ;  Total (Industry) ;.1",
                            "Turnover ;  Northern Territory ;  Total (Industry) ;.1",
                            "Turnover ;  Australian Capital Territory ;  Total (Industry) ;.1",
                            
                                                   ]]

#chose 
chosen_state_df = question_2_df.filter(regex= 'Month|')
```

```python
for state in state_list:

    chosen_state = state

    chosen_state_df = question_2_df.filter(regex= f'Month|{chosen_state}')

# Set 'Month' as index for chosen_state_df

    chosen_state_df = chosen_state_df.set_index('Month')

# From chosen_state_df, create a new dataframe contains only data from the last decade

    chosen_state_last_decade_df = chosen_state_df.loc['2013-01-01':'2023-12-31']

# Create a line chart showing the industry turnover by time for the chosen state in the last decade

    chosen_state_last_decade_df.plot(kind='line')

    plt.title(f"Industry Turnover by Month {chosen_state} - Seasonally Adjusted ")
    plt.xlabel('Month')
    plt.ylabel('Turnover ($ Millions)')
    plt.legend()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.07),
          fancybox=True, shadow=True, ncol=5)

plt.show()
```
