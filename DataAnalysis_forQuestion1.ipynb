```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
data_by_state =pd.read_excel('Data/Table 3. Retail turnover, by state 850103.xlsx',sheet_name="Data1")
data_by_state = data_by_state.drop(range(0,9))
data_by_state = data_by_state.rename(columns = {'Unnamed: 0':'Month'})
data_by_state.head()
```

```python
#Filters data from January 2013 to last date September 2023
data_by_state = data_by_state.drop(range(9,378))
data_by_state
```

```python
question_1  = data_by_state.iloc[:, 10:18]
question_1.insert(0,"Month",data_by_state['Month'],True)
question_1.head()
```

```python
question_1.plot(x='Month')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```

```python
population_data =pd.read_excel('Data/Australian Population data.xlsx',sheet_name="Data1")
population_data = population_data.drop(range(0,9))
population_data = population_data.rename(columns = {'Unnamed: 0':'Month'})
population_data.head()
```

```python
#Filters data from January 2013 to last date September 2023
population_data = population_data.drop(range(9,136))
population_data.head()
```

```python
state_population = population_data.iloc[:, 19:27]
state_population.insert(0,"Month",population_data['Month'],True)
state_population.head()
```

```python
state_population.plot(x='Month')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```

```python
latest_population = state_population.iloc[-1,1:9]
latest_population = pd.to_numeric(latest_population, errors='coerce')
latest_turnover = pd.to_numeric(question_1.iloc[-7,1:9], errors='coerce')
latest_turnover = latest_turnover*1000000

turnover_per_person =  latest_turnover.values / latest_population.values

states = ['NSW', 'VIC', 'QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
plt.bar(states,turnover_per_person, width = 0.5)
plt.ylim(1200,1450)
plt.xlabel('Australian States')
plt.ylabel("Retail Turnover per Person (in AUD)")
plt.title('Retail Turnover per person vs State')
plt.show()
```
